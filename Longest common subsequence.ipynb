```python
def read_fasta_file(file_path):
    sequences = []
    with open(file_path, 'r') as file:
        lines = file.readlines()
        sequence = ''
        for line in lines:
            if line.startswith('>'):
                if sequence:
                    sequences.append(sequence)
                    sequence = ''
            else:
                sequence += line.strip()
        if sequence:
            sequences.append(sequence)
    return sequences
```

#  Dynamic programming Approach

```python
def find_lcs(string1, string2):
    m = len(string1)
    n = len(string2)

    lcs = [[0] * (n+1) for _ in range(m+1)]

    for i in range(m):
        for j in range(n):
            if string1[i] == string2[j]:
                lcs[i+1][j+1] = lcs[i][j] + 1
            else:
                lcs[i+1][j+1] = max(lcs[i][j+1], lcs[i+1][j])

    # Backtracking to find LCS
    lcs_length = lcs[m][n]
    lcs_string = [''] * lcs_length
    i = m - 1
    j = n - 1

    while i >= 0 and j >= 0:
        if string1[i] == string2[j]:
            lcs_string[lcs_length - 1] = string1[i]
            i -= 1
            j -= 1
            lcs_length -= 1
        elif lcs[i][j+1] > lcs[i+1][j]:
            i -= 1
        else:
            j -= 1

    return ''.join(lcs_string)

# Example usage
string1 = "ABCDGH"
string2 = "AEDFHR"
result = find_lcs(string1, string2)
print(result)  # Output: ADH
```
